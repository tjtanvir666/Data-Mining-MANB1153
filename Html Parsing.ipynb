```python
r = 0
places = ["cyberjaya","puchong","kuala lumpur","shah alam",]
info_dict = {}
info_dit_main = {}


while r <= 1:
    add = 'https://openweathermap.org/data/2.5/find?callback=?&q='+places[r]+',my&type=like&sort=population&cnt=30&appid=b6907d289e10d714a6e88b30761fae22'
    print(places[r])
    data = requests.get(add).text
    data = data.strip("?()[]")
    data=json.loads(data)
    #info_dict[places[r]] = data
    #print(data["list"],"\n")
    
    info_1 = {}
    info_2 = {}
   
    for val in data["list"]:
        info_1 = {"Location"                      : val['name'],
                  "Latitude"                      : val["coord"]["lat"], 
                  "Longitude"                     : val["coord"]["lon"],
                  "Date"                          : datetime.fromtimestamp(val['dt']).strftime('%Y-%m-%d %H:%M:%S'),
                  "Clouds(%)"                     : val['clouds'],
                  "Temparature(in celcius)"       : (10*(val['main']["temp"] - 273.15))/10, 
                  "Max_temp"                      : (10*(val['main']["temp_max"] - 273.15))/10,
                  "Min_temp"                      : (10*(val['main']["temp_min"] - 273.15))/10, 
                  "Humidity(%)"                   : val['main']["humidity"],
                  "Pressure(hpa)"                 : val['main']["pressure"],
                  "Rain"                          : val["rain"],
                  "Wind(m/s)"                     : val["wind"]['speed'],
                  "Wind_Direction(in degree)"     : val["wind"]['deg']}
        
        #print("this is the iteration", val["main"], "\n")
        #pprint.pprint(info_1)
        
        for item in  val["weather"]:
            info_2 = {"Weather_Condition"   : item["main"],
                      "Weather_Description" : item["description"]}
       
        #pprint.pprint(info_2)
    
    for d in [info_1, info_2]:
        info_dict.update(d)
    info_dit_main[r] = info_dict            
       
    r+=1
    print(r,"\n")
pprint.pprint( info_dit_main)    
```
